## Line Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# linspace is a way of generating data starting from the 
# 'start point', 'end point', and how many data from the
# left to the right
x = np.linspace(0, 10, 10)
print(x)
print("*" * 50)
print("*" * 50)
y = np.sin(x)
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
```

```python
plt.xlabel("Time")
plt.ylabel("y-lable")
plt.title("this is the title")
plt.plot(x,y)
plt.show()
```

## Scatterplot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
A = pd.read_csv('data_1d.csv', header=None).as_matrix()
```

```python
x = A[:,0]
y = A[:,1]
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
np
```

```python
x_line = np.linspace(0,100,100)
y_line = 2*x_line + 1
plt.scatter(x, y)
plt.plot(x_line, y_line)
plt.show()
```

## Histogram

```python
plt.hist(x)
plt.show()
```

```python
R = np.random.random(10000)
plt.hist(R)
plt.show()
```

```python
# you can also control the number of bins
plt.hist(R, bins=10)
plt.show()
```

```python
y_actual = 2*x + 1
residuals = y - y_actual
plt.hist(residuals)
plt.show()
```
